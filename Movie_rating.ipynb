```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
import warnings
warnings.simplefilter("ignore")
```

```python
from sklearn.preprocessing import scale
from sklearn.manifold import TSNE
from sklearn.cluster import KMeans
```

```python
ds = pd.read_csv('/kaggle/input/movie-rating/IMDb Movies India.csv',encoding='latin-1')
ds.head()
```

```python
ds.shape
```

```python
year = []
for i in ds.Year:
    if type(i) == float:
        year.append(np.nan)
    else:
        year.append(int(str(i)[1:5]))
ds["Year"] = year
ds.head()
```

```python
duration = []
for k in ds.Duration:
    if type(k) == float:
        duration.append(np.nan)
    else:
        duration.append(int(str(k).split(" ")[0]))
ds["Duration"] = duration
ds.head()
```

```python
ds["Votes"]=ds["Votes"].replace("$5.16M",516)
ds["Votes"] = pd.to_numeric(ds['Votes'].str.replace(',',''))
ds.head()
```

```python
ds.info()
```

```python
ds.describe()
```

```python
ds.isnull().sum()
```

```python
def TopTen(column):
    global ds
    ds[column].value_counts().sort_values(ascending=False)[:10].plot(kind="bar", figsize=(20,6), edgecolor="k")
    plt.xticks(rotation=0)
    plt.title("Top Ten {}".format(column))
    plt.xlabel(column)
    plt.ylabel("Count")
    plt.show()
```

```python
TopTen("Actor 3")
```

```python
def Histo(column):
    global ds
    plt.figure(figsize=(20,6))
    plt.hist(ds[column], edgecolor="k")
    plt.xticks(rotation=0)
    plt.title("Histogram of {}".format(column))
    plt.xlabel(column)
    plt.ylabel("Frequency")
    plt.show()
```

```python
top = ds.sort_values("Votes", ascending=False).head(10)[[ "Name", "Year", "Duration", "Votes", "Director","Actor 1"]].reset_index(drop=True)
```

```python
top
```

```python
worst = ds.sort_values("Votes", ascending=False).dropna().tail(10)[[ "Name", "Year", "Duration", "Votes", "Director","Actor 1"]].reset_index(drop=True)
```

```python
worst
```

```python
plt.figure(figsize=(20,6),dpi=100)
plt.scatter(top["Year"], top["Duration"], edgecolor="k",s=150, label="Top 10")
for j in range(10):
    plt.text(x=top["Year"][j]+0.8,y=top["Duration"][j]-1.2,s=top["Name"][j], fontsize=11, rotation =0)
plt.scatter(worst["Year"], worst["Duration"], color="red",s=150, label="Worst 10", edgecolor="k")
for j in range(10):
    plt.text(x=worst["Year"][j]-1.8,y=worst["Duration"][j]+3,s=worst["Name"][j], fontsize=11, rotation =0)
plt.legend()
plt.title("The 10 movies with the most votes (with Movie Name)", fontsize=20)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Duration", fontsize=12)
plt.xticks(rotation=0)
plt.show()
```
